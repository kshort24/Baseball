# Inning Level Aggregate

```python
def summarize_inning(df):
    x = df.groupby(['game_pk','inning'],as_index=False
          ).agg({'launch_speed' : 'sum'
                 ,'des' : 'count'
                 ,'at_bat_number' : 'nunique'
                 ,'estimated_ba_using_speedangle' : 'sum'
                 ,'estimated_woba_using_speedangle' : 'sum'
                 ,'inning_topbot' : 'max'
                 ,'home_team' : 'max'
                 ,'post_home_score' : 'max'
                 ,'home_score' : 'min'
                 ,'away_team' : 'max'
                 ,'post_away_score' : 'max'
                 ,'away_score' : 'max'
                 ,'game_date' : 'max'
                 #,'game_year' : 'max'
                 # ,'fielder_2' : 'max'
                 # ,'fielder_3' : 'max'
                 # ,'fielder_4' : 'max'
                 # ,'fielder_5' : 'max'
                 # ,'fielder_6' : 'max'
                 # ,'fielder_7' : 'max'
                 # ,'fielder_8' : 'max'
                 # ,'fielder_9' : 'max'
                 # ,'pitcher' : 'max'
                })
    return x
```
