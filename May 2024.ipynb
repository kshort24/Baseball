```python
def game_summary(df):
    x = df.groupby('game_date', as_index=False
              ).agg({'game_pk' : 'max'
                     ,'home_team' : 'max'
                     ,'post_home_score' : 'max'
                     ,'away_team' : 'max'
                     ,'post_away_score' : 'max'
                     ,'estimated_ba_using_speedangle' : 'mean'
                     ,'estimated_woba_using_speedangle' : 'mean'
                     ,'des' : 'count'
                     ,'pitcher' : 'nunique'
                     ,'inning' : 'nunique'
                     ,'at_bat_number' : 'nunique'
                    }).round(3)
    y = results('home_team')
    gs = x.merge(y, left_on = 'home_team', right_on = 'home_team', suffixes = ('','_y'))
    return gs

game_summary(df)
```
